```python
import matplotlib.pyplot as plt

each_value = []
probabilities = []
def_prob = 1

for value in range(1, 51):
    each_value.append((365 - value) / 365)
    def_prob *= each_value[-1]
    probabilities.append(def_prob)


plt.plot(range(1, 51), probabilities, marker='o', label='Probability')
plt.axhline(y=0.5, color='r', linestyle='--', label='0.5 Probability')

plt.title('Probability of Shared Birthdays')
plt.xlabel('Number of People')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
```
