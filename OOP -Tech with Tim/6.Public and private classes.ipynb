### Intro:


```python
# private can only be accessed inside the class
class _Private:
  def __init__(self, name):
    self.name = name


# Public classes are visible to other classes.
# This means that any other class can access a
# public field or method.
# Furthermore, other class can modify a public field
# unless the field is declared final.
class NotPrivate:
  def __init__(self, name):
    self.name = name
    self.priv = _Private(name)
  
  def _display(self):
    print("Hello")
  
  def display(self):
    print("Hi")
```

The above definition is for java.
On python there is no such thing as public or private classes.
There are conventions we used to declare public or private class but there is no restriction with any classes

```python
# usually pre _ denotes private classes
# in a private class, every method and attributes are to be treated as the same
class _Private:
  def __init__(self, name):
    self.name = name

class NotPrivate:
  def __init__(self, name):
    self.name = name
    self.priv = _Private(name)
  
  # this is telling other people or myself to not use this method
  # thus keeping it as a private method
  def _display(self):
    print("Hello")
  
  def display(self):
    print("Hi")
```

* This is useful when learning other languages.
* "_" prefix is also a general convention indicating private classes
