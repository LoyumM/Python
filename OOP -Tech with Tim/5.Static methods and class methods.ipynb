## Intro:

```python
class Dog:
  # this is a class varible
  # written outside any methods
  # and usually at the top of the class
  # variables that will be referenced in multiple method are put here
  dogs = []

  def __init__(self, name):
    self.name =  name
    self.dogs.append(self)

  @classmethod
  def num_dogs(cls):
    return len(cls.dogs)

  @staticmethod
  def bark(n):
    """bark n times"""
    for i in range(n):
      print("bark!")

tim = Dog("Aby")
jim = Dog("Bony")
# returned as pointers but the two names are still returned
print(Dog.dogs)
```

## decorators:

```python
class Dog:
  dogs = []

  def __init__(self, name):
    self.name =  name
    self.dogs.append(self)

  @classmethod
  # "@random_variable" is decorators
  # put them above your method if you want to indicate that
  # they are a different type of method
  def num_dogs(cls):
    return len(cls.dogs)

  @staticmethod
  # another decorator indicating that it is a static methof
  def bark(n):
    """bark n times"""
    for i in range(n):
      print("bark!")
```

## class methods:

```python
class Dog:
  dogs = []

  def __init__(self, name):
    self.name =  name
    self.dogs.append(self)

  @classmethod
  # you can call a class method with the name of the class
  # there is no self here
  def num_dogs(cls):
    return len(cls.dogs)

  @staticmethod
  def bark(n):
    """bark n times"""
    for i in range(n):
      print("bark!")

# this works
print(Dog.num_dogs())
# and you don't need to do
tim = Dog("Aby")
print(tim.num_dogs())
```

## static methods

```python
class Dog:
  dogs = []

  def __init__(self, name):
    self.name =  name
    self.dogs.append(self)

  @classmethod
  def num_dogs(cls):
    return len(cls.dogs)

  @staticmethod
  # don't need the class to be called
  # doesn't pass in the class
  # we don't need self, class or any parameter
  # "n" parameter is not mandatory
  def bark(n):
    """bark n times"""
    for i in range(n):
      print("bark!")

# notice that this doesn't touch any attributes inside the class
# and there is no way to do that
Dog.bark(10)
```

## Use of static methods:
* they are used as functions organised as classes
* Why create classes when they are used as functions?
This is done so that they can be made as importable modules in python

```python
## example:

class Math:
  @staticmethod
  def add(x, x2):
    return x + x2
  @staticmethod
  def sub(x, x2):
    return x - x2
```

* Class methods are more used for 
accessing class variables without having to pass in an object.
* Decorators are used for telling python and users that it is not a regular method, so the methods doesn't need to have a self or a paramters.
* Multiple numbers of classmethods and static methods can be created
