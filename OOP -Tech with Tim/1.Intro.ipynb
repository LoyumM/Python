### What is objects?

```python
# here x is an instance of int class and the value is 5
x = 5
# similarly y is also an instance of str class and value is 'string'
y = 'string'

print(type(x))
print(type(y))
```

```python
# classes can have built in functionality
print(x/4) 
# upper(), lower(), strip(), etc. are all built in method of str class
print(y.upper())
```

```python
# this throws an error because int class doesn't have strip() method built in
x.strip()
```
