## Inheritence

```python
# if you notice the two classes are mostly similar. 
# Inheritence is used to avoid copy pasting similar mmethods and attributes
class Dog(object): 

  def __init__(self, name, age):
    self.name = name
    self.age = age
  
  def speak(self):
    print("Hi, I am", self.name, " and is", self.age, "years old.")

class Cat(object):
  def __init__(self, name, age, color):
    self.name = name
    self.age = age
    self.color = color
  
  def speak(self):
    print("Meow! my name is ", self.name, ". I am", self.age, "years old and I'm ", self.color, "in color.")

tim = Cat('Aki', 6, 'Black')
tim.speak()
```

### Parent and child

```python
## dog is the parent or super class
class Dog(object): 

  def __init__(self, name, age):
    self.name = name
    self.age = age
  
  def speak(self):
    print("Hi, I am", self.name, " and is", self.age, "years old.")

## Cat is the child or derived class 
# Cat is going to inherit from dog
# Cat inherits all of the properties, methods and attributes from Dog
class Cat(Dog):

  def __init__(self, name, age, color):
    super().__init__(name,age)
    # takes the constructor(__init__) from super().i.e., Dog
    # very useful in long methods and classes
    self.color = color

tim = Cat('Aby', 5, 'blue')
tim.speak()
```

### Overloading or overwritting

```python
class Dog(object): 

  def __init__(self, name, age):
    self.name = name
    self.age = age
  
  def speak(self):
    print("Hi, I am", self.name, "and I am", self.age, "years old.")

  def talk(self):
    print("Bark")

class Cat(Dog):

  def __init__(self, name, age, color):
    super().__init__(name,age)
    self.color = color

  ## overloading or overwritting 
  def talk(self):
    # just because
    for i in range(10):
      print("Meow Meow!")


jim = Dog('Tom', 11)
jim.speak()
jim.talk()


tim = Cat('Aby', 5, 'blue')
tim.speak()
tim.talk()
```
