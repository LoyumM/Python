## Example of overloading/overriding

Here is an example of overloading default classes on python

```python
class Point():
  def __init__(self, x=0, y=0):
    self.x = x
    self.y = y
    self.coords = (self.x, self.y)

  def move(self, x, y):
    self.x += x
    self.y += y

  def __add__(self, p):
    return Point(self.x + p.x, self.y + p.y)

  def __sub__(self, p):
    return Point(self.x - p.x, self.y - p.y)


  def __mul__(self, p):
    return self.x * p.x + self.y * p.y

p1 = Point(3, 4)
p2 = Point(3, 2)
p3 = Point(1, 3)
p4 = Point(0, 1)
p5 = p1 + p2
p6 = p4 - p1
p7 = p2*p3

# returns pointers instead?? No idea why. look up later
# i.e., the address in memory
print(p5, p6, p7)
```

```python
class Point():
  def __init__(self, x=0, y=0):
    self.x = x
    self.y = y
    self.coords = (self.x, self.y)

  def move(self, x, y):
    self.x += x
    self.y += y

  def __add__(self, p):
    return Point(self.x + p.x, self.y + p.y)

  def __sub__(self, p):
    return Point(self.x - p.x, self.y - p.y)

  def __mul__(self, p):
    return self.x * p.x + self.y * p.y

  # to return the pointers as strings
  def __str__(self):
    return "(" +str(self.x) + "," + str(self.y) + ")"

p1 = Point(3, 4)
p2 = Point(3, 2)
p3 = Point(1, 3)
p4 = Point(0, 1)
p5 = p1 + p2
p6 = p4 - p1
p7 = p2*p3

# only works when returned as string
print(p5, p6, p7)
```



*   the functions gets triggered without  calling them because they are built in functions
*   here in this tutorials they were only redefined to do something else



## Comparisions:

```python
import math
class Point():
  def __init__(self, x=0, y=0):
    self.x = x
    self.y = y
    self.coords = (self.x, self.y)

  def move(self, x, y):
    self.x += x
    self.y += y

  def __add__(self, p):
    return Point(self.x + p.x, self.y + p.y)

  def __sub__(self, p):
    return Point(self.x - p.x, self.y - p.y)

  def __mul__(self, p):
    return self.x * p.x + self.y * p.y

  def length(self):
    # pythogorean theorem(a**2 + b**2 == c**2)
    return math.sqrt(self.x**2 + self.y**2)
  
  def __gt__(self,p):
    return self.length() > p.length()

  def __ge__(self,p):
    return self.length() >= p.length()

  def __lt__(self,p):
    return self.length() < p.length()

  def __le__(self,p):
    return self.length() <= p.length()

  # not exactly true
  def __eq__(self,p):
    return self.x == p.x and self.y == p.y

  def __str__(self):
    return "(" +str(self.x) + "," + str(self.y) + ")"

p1 = Point(3, 4)
p2 = Point(3, 2)
p3 = Point(1, 3)
p4 = Point(0, 1)


print(p1 == p2)
print(p2 > p4)
print(p2 <= p3)
```
