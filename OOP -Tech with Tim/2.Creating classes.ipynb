## Class and methods


```python
# dog is the class
class Dog(object):
## __init__ is what is known as the constructor method
## __init__ will automotically run every time an object of type Dog  
  def __init__(self):
    print("Nice you made a Dog!")
# the function are called methods  
# you can create as many methods as you want
  def speak(self):
    pass

tim = Dog()
```

## Instance and attrbutes

```python
class Dog(object): 
# self defines the instance you're calling
  def __init__(self, name):
    self.name = name
  
  def speak(self):
    print("Hi, I am", self.name)
# tim is what is being passed into self
# so it's saying tim.name(self.name) is whatever name we assign here
tim = Dog('Aby')
# similarly fred is passed to self
fred = Dog("Bob")
tim.speak()
fred.speak()
```

```python
## more examples of instances
class Dog(object): 

  def __init__(self, name, age):
    # name and age are attributes
    self.name = name
    self.age = age
  
  def speak(self):
    print("Hi, I am", self.name, " and is", self.age, "years old.")

  def change_age(self, age):
    ## self.age is an attribute
    self.age = age

tim = Dog('Aby', 55)
fred = Dog("Bob", 6)
tim.change_age(5)
tim.speak()
fred.speak()

## you can also print just the attributes like age or name
# example:
print(tim.age)
print(fred.name)
```
